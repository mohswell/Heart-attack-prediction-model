```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("heart.csv")
df.head(4)
```

# Splitting the Data into X and Y respectively in ratio of 3 : 1

```python
X = df.drop (columns="target", axis=1)
Y = df["target"]
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print (X.shape, X_train.shape, X_test.shape)
```

```python
print (Y.shape, Y_train.shape, Y_test.shape)
```

```python
#IN THE ABOVE CODE: IT SHOWS 80% OF THE DATA is for Training Y based on columns in X
# AND remaining 20% is for testing the model based on X and Y
```

# Training the Model using Logistic Regression

```python
LogisticRegression(solver='lbfgs', max_iter=10000)
```

```python
model = LogisticRegression()

# Training the LogisticRegression model with the Training data
model.fit (X_train.values, Y_train)
```

# Evaluating The Model By Predicting Values (0 or 1)

```python
# Predicting the output with test data
y_predict = model.predict (X_test.values)
print (y_predict)
```

```python
#Printing the accuracy percentge of the model 
accuracy_result = round(100 * accuracy_score(Y_test,y_predict), 4)
print(accuracy_result)
```

```python
if (y_predict[0]== 1):  
    print ("The Person does not have a Heart Disease")
else:  
    print ("The Person has Heart Disease")
```

# Making 2nd party Predictions based on this algorithm

```python
input_data = (62, 0, 0, 140, 268, 0, 0, 160, 0, 3.6, 0, 2, 2)

numpy_data= np.asarray (input_data)
# reshape the numpy array as we are predicting for only one instance
input_reshaped = numpy_data.reshape (1,-1)
prediction = model.predict (input_reshaped)

if (prediction[0]== 0):  
    print ("The Person does not have a Heart Disease")
else:  
    print ("The Person has Heart Disease")
```
